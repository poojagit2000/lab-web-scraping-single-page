```python
# You have to find data on the internet about currently popular songs. Popvortex maintains a weekly Top 100 of "hot" songs here: http://www.popvortex.com/music/charts/top-100-songs.php.

# It's a good place to start! Scrape the current top 100 songs and their respective artists, and put the information into a pandas dataframe.
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import random
```

```python
url = "http://www.popvortex.com/music/charts/top-100-songs.php"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
soup.select("#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p")
```

```python
song_details = soup.find_all(class_ = "title-artist")
song_details
song_names = []
artist_names = []
for rows in song_details:
    song_name = rows.find(class_ = "title")
    artist_name =rows.find(class_ = "artist")
    song_names.append(song_name.text.strip())
    artist_names.append(artist_name.text.strip())
```

```python
song_details
```

```python
df = pd.DataFrame({'Title': song_names, 'Artist': artist_names})
```

```python
df
```

```python
user_input = input("Enter the song name for the recommendation :")
list_Title = df.Title.values
if user_input in list_Title:
    print(random.choice(list_Title))
else:
    print("Cannot recommend the song")
    
```
